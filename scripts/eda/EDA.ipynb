## **3. Distribucion de variables**
---

```python
# Histograma de las variables numéricas
df_insuficiencia_cardiaca.hist(bins=15, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle("Distribución de variables numéricas", fontsize=16)
plt.tight_layout()
plt.show()
```

## **4. Correlacion Variables numericas**
---

```python
# Matriz de correlación
correlation_matrix = df_insuficiencia_cardiaca.corr()

# Mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```

## **5. Exploracion Variable objetivo**
---

```python
# Distribución de la variable objetivo
sns.countplot(data=df_insuficiencia_cardiaca, x='DEATH_EVENT', palette='pastel')
plt.title("Distribución de la variable objetivo (DEATH_EVENT)")
plt.show()

# Comparación de variables por la variable objetivo
for column in df_insuficiencia_cardiaca.columns:
    if column != 'DEATH_EVENT':
        sns.boxplot(data=df_insuficiencia_cardiaca, x='DEATH_EVENT', y=column, palette='Set3')
        plt.title(f"Distribución de {column} por DEATH_EVENT")
        plt.show()
```

## **6. Identificación de valores atípicos**
---


```python
# Usar diagramas de caja (boxplots) para identificar valores atípicos
numeric_columns = df_insuficiencia_cardiaca.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_columns:
    sns.boxplot(data=df_insuficiencia_cardiaca, x=column, color='skyblue')
    plt.title(f"Diagrama de caja para {column}")
    plt.show()
```
